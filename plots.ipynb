```python
import matplotlib.pyplot as plt
import numpy as np
```

## MergeSmall_k

```python
sizes = [16, 32, 64, 128, 256, 512, 1024]
#Average execution times in microseconds (Average over 1000000 runs)
exec_times = [4, 5, 5, 6, 7, 8, 9]
```

```python
plt.figure(figsize=[10,5])
plt.plot(sizes, exec_times)
plt.title("Temps d'exécution de MergeSmall_k")
plt.xlabel("|A|+|B|")
plt.ylabel("Temps d'exécution")
plt.show()
```
